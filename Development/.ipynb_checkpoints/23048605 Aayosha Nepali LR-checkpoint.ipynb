# Model Training
## Linear Regression
### Setting Up Feature and Target

```python
features = ['season', 
            'round', 
            'grid', 
            'lat', 
            'long', 
            'driverExperience', 
            'teamStrength',
    'driverAge']
# y = position
```

### Train Test Split

```python
X = results[features]
y = results['position']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))
```

```python
X_train_scaled.shape, X_test.shape, y_train_scaled.shape, y_test.shape
```

```python
y_train_scaled
```

### Convert to PyTorch Tensors

```python
X_train_t = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_t = torch.tensor(y_train_scaled, dtype=torch.float32)

X_test_t = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_t = torch.tensor(y_test_scaled, dtype=torch.float32)
```

## Initialise Weights and Bias

```python
w = torch.randn(X_train_t.shape[1], 1)
b = torch.randn(1)
```

```python
w.shape, b.shape
```

```python
epochs = 1100
lr = 0.01
loss_history = []

for epoch in range(epochs):
    y_pred = X_train_t @ w + b
    error = y_pred - y_train_t
    loss = torch.mean(error ** 2)

    # Gradients
    dw = (2 / len(y_train_t)) * (X_train_t.T @ error)
    db = 2 * torch.mean(error)

    # Update
    w -= lr * dw
    b -= lr * db

    loss_history.append(loss.item())

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

### Loss Curve

```python
plt.figure(figsize = (8, 5))
plt.plot(loss_history)
plt.title("Training Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

## Evaluation on Test Data

```python
with torch.no_grad():
    y_test_pred_scaled = X_test_t @ w + b

y_test_pred = scaler_y.inverse_transform(
    y_test_pred_scaled.numpy()
)

y_test_original = scaler_y.inverse_transform(
    y_test_t.numpy()
)
```

## Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test_original, y_test_pred)
rmse = np.sqrt(mean_squared_error(y_test_original, y_test_pred))
r2 = r2_score(y_test_original, y_test_pred)

print(f"MAE  : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.3f}")
```

## Train Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha = 1.0, random_state = 42)
ridge.fit(X_train_scaled, y_train)
```

### Evaluate Ridge Regression

## Creating the New Feature Matrix

```python
num_features = ['season', 'round', 'grid', 'lat', 'long', 'driverExperience', 'teamStrength',
    'driverAge']
```

### One‑Hot Encoding

```python
X_cat = pd.get_dummies(
    results['experience_group'],
    drop_first=True   # avoids dummy variable trap
)
```

```python
# combining numerical and categorical
X_num = results[num_features]

X_full = pd.concat([X_num, X_cat], axis=1)
y = results['position']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_full, y, test_size = 0.3, random_state = 42)
```

### Scaling Numerical and Dummy Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Training Linear Regression With New Features

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

lr_new = LinearRegression()
lr_new.fit(X_train_scaled, y_train)

y_pred_lr_new = lr_new.predict(X_test_scaled)
r2_lr_new = r2_score(y_test, y_pred_lr_new)

print("Linear Regression with experience_group")
print("MAE :", mean_absolute_error(y_test, y_pred_lr_new))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr_new)))
print("R²  :", r2_score(y_test, y_pred_lr_new))
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

param_grid = {'alpha': [0.1, 1, 10, 50, 100]}

ridge = Ridge(random_state = 42)

grid_ridge = GridSearchCV(
    ridge,
    param_grid,
    scoring = 'neg_mean_squared_error',
    cv = 5
)

grid_ridge.fit(X_train_scaled, y_train)

best_ridge = grid_ridge.best_estimator_
y_pred_ridge_new = best_ridge.predict(X_test_scaled)
r2_ridge_new = r2_score(y_test, y_pred_ridge_new)

print("Ridge with experience_group")
print("Best alpha:", grid_ridge.best_params_['alpha'])
print("MAE :", mean_absolute_error(y_test, y_pred_ridge_new))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_ridge_new)))
print("R²  :", r2_score(y_test, y_pred_ridge_new))
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso

param_grid = {'alpha': [0.001, 0.01, 0.1, 1]}

lasso = Lasso(max_iter = 5000, random_state = 42)

grid_lasso = GridSearchCV(
    lasso,
    param_grid,
    scoring = 'neg_mean_squared_error',
    cv = 5
)

grid_lasso.fit(X_train_scaled, y_train)

best_lasso = grid_lasso.best_estimator_
y_pred_lasso_new = best_lasso.predict(X_test_scaled)
r2_lasso_new = r2_score(y_test, y_pred_lasso_new)

print("Lasso with experience_group")
print("Best alpha:", grid_lasso.best_params_['alpha'])
print("MAE :", mean_absolute_error(y_test, y_pred_lasso_new))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lasso_new)))
print("R²  :", r2_score(y_test, y_pred_lasso_new))
```
